```python
import pandas as pd   #to create dataframe
import requests       #to send the request to the URL
from bs4 import BeautifulSoup #to get the content in the form of HTML
import numpy as np 
```

```python
url = 'https://www.imdb.com/search/title/?title_type=feature&genres=adventure&release_date=2024-01-01,2024-12-31'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#print(soup)
```

```python
movie_name = []
genre =[]
Duration = []
rating = []
votes = []
```

```python
movie_data = soup.findAll('div', attrs= {'class':"ipc-metadata-list-summary-item__tc"})
```

```python
for store in movie_data:
    name = store.h3.text
    movie_name.append(name)

    runtime_1 = store.div.find_all('span', class_ = 'sc-d5ea4b9d-7 URyjV dli-title-metadata-item')
    runtime_text = runtime_1[1].text.replace(' min', '')
    hours = 0
    minutes = 0

    if 'h' in runtime_text:  # Check if hours are present
        hours = int(runtime_text.split('h')[0])  # Get the number of hours

    if 'm' in runtime_text:  # Check if minutes are present
        minutes = int(runtime_text.split('h')[-1].replace('m', '').strip())  # Get the number of minutes

# Convert hours to minutes and add the remaining minutes
    total_minutes = hours * 60 + minutes

    Duration.append(total_minutes)
    rate = store.div.find_all('span', class_ = 'ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating')
     # Split by space and take the first part (e.g., '6.1')
    for item in rate:
    # Extract IMDb rating (e.g., 6.6)
        rating_value = item.find('span', class_='ipc-rating-star--rating').text.strip()
        rating.append(rating_value)
        
    # Extract vote count (e.g., 195K)
        vote_count = item.find('span', class_='ipc-rating-star--voteCount').text.strip()
        votes.append(vote_count)

    genre_1 = store.find('ul', class_='ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--no-wrap ipc-inline-list--inline baseAlt')
    
    # Check if genre exists and extract them
    if genre_1:
        genre_list = [genre.text.strip() for genre in genre_1.find_all('li')]
        genre.append(', '.join(genre_list))  # Join multiple genres with a comma
    
```

```python
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Genre' : "Adventure", 'Duration': Duration, 'Ratings': rating, 'Votes Count': votes})
```
